```python
import re
import urllib
import urllib2
import json
import string

import nltk


def getJSON(url):
    response = urllib.urlopen(url)
    data = json.loads(response.read())
    return data


WATSON_API_KEY = 'INSERT-HERE'
```

```python
text = '''
Responding at Wednesday’s Prime Minister’s Questions to Jeremy Corbyn’s 
visit to the refugee camp at Calais last weekend, Cameron joked: “They 
met with a bunch of migrants in Calais. They said they could all come to Britain.” 
Corbyn had travelled to Calais to witness for himself the appalling conditions 
endured by those at the camp, by people who have already fled war or extreme 
poverty, and survived treacherous journeys across Europe and the Mediterranean. 
Britain and France need to be “a bit more human about this”, Corbyn said.
'''

text = text.replace('‘','"').replace('’','"').replace('“','"').replace('”','"').replace(',',',')

print text
```

```python
def getWatsonSentiment(s):
    '''
    Function to get the sentiment of a string using the IBM Watson API (Alchemy).
    API KEY needs to be specified.
    '''
    text = s.translate(string.maketrans("",""), string.punctuation)
    # text = s.encode('utf8').translate(string.maketrans("",""), string.punctuation)
    
    watsonURL = 'http://gateway-a.watsonplatform.net/calls/text/TextGetTargetedSentiment'
    watsonURL += '?apikey=' + WATSON_API_KEY
    watsonURL += '&target=' + 'corbyn' + '&text=' + text
    watsonURL += '&outputMode=json'
    watson_data = getJSON(watsonURL)
    
    return watson_data


print json.dumps(getWatsonSentiment(s=text),indent=4)
```
