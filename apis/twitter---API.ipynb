```python
def getTweetsForQuery(query, input_tweets=None):
    
    if input_tweets is None:
        output = []
    else:
        output = input_tweets
    
    for i in range(180):
        if i % 10 == 0:
            print i, '.', 
        if i == 0:
            prev_id = '999998401048145920'
        search_results = twitter_api.search.tweets(q=query, count=100, max_id=prev_id)
        prev_id = search_results['statuses'][-1]['id']
        output += search_results['statuses']
    
    return output
```

```python
terms = ['syrianrefugees', 'SyrianRefugees', 'Syrianrefugees', 'Refugees', 'refugees', 'greece', 'RefugeeCrisis']
```

```python
tweets = getTweetsForQuery(query='#RefugeeCrisis', input_tweets=tweets)
print '\n\n', len(tweets)
```

```python
hashtags_collection = dict()

for tweet in tweets:
    for hashtag in tweet['entities']['hashtags']:
        if hashtag['text'] not in hashtags_collection:
            hashtags_collection[hashtag['text']] = [tweet['id']]
        else:
            hashtags_collection[hashtag['text']].append(tweet['id'])
```
